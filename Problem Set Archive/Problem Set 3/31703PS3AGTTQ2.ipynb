```python
np.random.seed(101)
N_obs = 100
```

```python
data = dict()
data['X'] = np.random.normal(size=N_obs)
data['Y'] = data['X'] ** 2 + np.random.normal(size=N_obs)
```

```python
data = pd.DataFrame(data)
```

```python
fig1, ax1 = plt.subplots(figsize=(5, 5), dpi=600)

ax1.scatter(data['X'], data['Y'], s=4);
ax1.set_xlabel(r'$X$', fontsize=13);
ax1.set_ylabel(r'$Y$', fontsize=13);
ax1.set_title('Plot of Data', fontsize=15);
```

We have defined a Class `Tree` that implements the required functions as methods. The structure of the class bypasses the need to return an output, as the methods work by updating the object's node attributes with new objects of the same class itself.

### Part A: Terminal Node Function

Please refer to the method `createTerminalNode`.

### Part B: Left and Right Node Functions

Please refer to the methods `addLeftNode` and `addRightNode`.

### Part C: Example

```python
tree_ex = Tree()
tree_ex.createTerminalNode(data=data, index=np.arange(0,6))
tree_ex.addLeftNode(data=data, index=np.arange(0,4))
tree_ex.addRightNode(data=data, index=np.arange(4,6))
tree_ex.left.addLeftNode(data=data, index=np.arange(0,2))
tree_ex.left.addRightNode(data=data, index=np.arange(2,4))

print('tree = {}'.format(tree_ex.index))
print('tree_left = {}'.format(tree_ex.left.index))
print('tree_left_left = {}'.format(tree_ex.left.left.index))
print('tree_left_right = {}'.format(tree_ex.left.right.index))
print('tree_right = {}'.format(tree_ex.right.index))
```

### Part D: Growing the Tree

Please refer to the method `growTree`.

### Part E: Prediction using the Tree

Please refer to the method `predictTree`.

### Part F: Prediction using Three-Layer Tree

```python
tree_ex = Tree()
tree_ex.createTerminalNode(data=data, index=data.index)
tree_ex.growTree(data=data)
tree_ex.left.growTree(data=data)
tree_ex.right.growTree(data=data)
tree_ex.left.left.growTree(data=data)
tree_ex.left.right.growTree(data=data)
tree_ex.right.left.growTree(data=data)
tree_ex.right.right.growTree(data=data)
```

```python
print('Number of unique clusters = {}'.format(np.unique(np.vectorize(tree_ex.predictTree)(data.index)).size))
```

```python
fig2, ax2 = plt.subplots(figsize=(5, 5), dpi=600)

ax2.scatter(data['X'], data['Y'], s=4);
ax2.scatter(data['X'], np.vectorize(tree_ex.predictTree)(data.index), s=10);
ax2.set_xlabel(r'$X$', fontsize=13);
ax2.set_ylabel(r'$Y$ (Predicted vs Actual)', fontsize=13);
ax2.set_title('Predicted vs Actual Data', fontsize=15);
```
